```python
# Program 1 
inputs = [1.2, 5.1, 2.1]
weights = [3.1, 2.1, 8.7]
bias = 3

output = inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + bias

print(output)
```

```python
# Program 2
inputs = [1, 2, 3] # input layer : these values could be values from 
# differnt sensors, or could be outputs from neuron also.
weights = [0.2, 0.8, -0.5] # Layer 1 
bias = 2

output = inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + bias

print(output)
```

For every neuron we have only 1 bias

```python
## adding one more intput to our program 2
inputs = [1, 2, 3, 2.5] # input layer
weights = [0.2, 0.8, -0.5, 1.0] # Layer 1 
bias = 2 # bias still remain 1. As we are dealing with 1 neuron only

output = inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + inputs[3]*weights[3] + bias

print(output)
```

```python
## Now here we will create a whole layer, with multiple neuron (4)

inputs = [1, 2, 3, 2.5] # input layer
weights1 = [0.2, 0.8, -0.5, 1.0] # Layer 1 neuron 1
weights2 = [0.5, -0.91, 0.26, -0.5] # Layer 1 neuron 2
weights3 = [-0.26, -0.27, 0.17, 0.87] # Layer 1 neuron 3

bias1 = 2 # Layer 1 neuron 1
bias2 = 3 # Layer 1 neuron 2
bias3 = 0.5 # Layer 1 neuron 3

output = [inputs[0]*weights1[0] + inputs[1]*weights1[1] + inputs[2]*weights1[2] + inputs[3]*weights1[3] + bias1,
            inputs[0]*weights2[0] + inputs[1]*weights2[1] + inputs[2]*weights2[2] + inputs[3]*weights2[3] + bias2,
            inputs[0]*weights3[0] + inputs[1]*weights3[1] + inputs[2]*weights3[2] + inputs[3]*weights3[3] + bias3]
print(output)
## we will have 3 outputs for 3 neurons each.
```
