# Topic 2. Basic Plot Types
Description: Cover basic plot types including line plots, scatter plots, bar plots, and histograms using Matplotlib.

### Example Code:

```python
import matplotlib.pyplot as plt
import numpy as np

# Line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='sin(x)')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()

# Scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='blue')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Bar plot
labels = ['A', 'B', 'C']
values = [3, 7, 2]
plt.bar(labels, values, color='green')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30, color='orange', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# Topic 3. Advanced Plot Types
Description: Introduce advanced plot types like box plots, violin plots, and heatmap using Matplotlib.

### Example Code:

```python
import matplotlib.pyplot as plt
import numpy as np

# Box plot
data = np.random.normal(size=(10, 4))
plt.boxplot(data, labels=['A', 'B', 'C', 'D'])
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# Violin plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.violinplot(data)
plt.title('Violin Plot')
plt.xlabel('Data')
plt.ylabel('Value')
plt.show()

# Heatmap
data = np.random.rand(10, 12)
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Heatmap')
plt.show()
```

# Topic 4. Subplots and Layouts
Description: Demonstrate creating multiple plots in a single figure using subplots and adjusting subplot arrangements.

### Example Code:

```python
import matplotlib.pyplot as plt
import numpy as np

# Subplots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].plot(x, y1, 'r-')
axs[0].set_title('Sin(x)')
axs[1].plot(x, y2, 'g-')
axs[1].set_title('Cos(x)')
plt.tight_layout()
plt.show()
```

# Topic 5. Adding Annotations and Text
Description: Teach adding titles, labels, legends, and annotations to plots for enhancing clarity and presentation.

### Example Code:

```python
import matplotlib.pyplot as plt
import numpy as np

# Adding annotations
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='sin(x)')
plt.title('Line Plot with Annotation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(np.pi/2+1, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend()
plt.show()
```

### More detail about the annotate method :

plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(np.pi/2+1, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))

- 'Peak' is the text of the annotation.
- xy=(np.pi/2, 1) specifies the point on the plot where the annotation arrow will point to. Here, it's at (/2, 1).
- xytext=(np.pi/2+1, 1.5) specifies the position of the text 'Peak'. It's placed slightly to the right and above the point (/2, 1).
- arrowprops=dict(facecolor='black', shrink=0.05) customizes the properties of the arrow pointing to the annotated point. facecolor sets the color of the arrow, and shrink controls the size of the arrow head relative to the arrow length.

For better understanding, try changint he values of the parameters inside the annotate method and see how it works.
