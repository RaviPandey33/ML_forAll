# What is linear regression and where is it used. 
Imagine you have a dataset where you know the X values (like hours spent studying) and the Y values (like test scores). Linear regression helps us find a line (a straight line) that predicts Y (test scores) based on X (hours studying). This line is drawn through the points on a scatter plot in such a way that it's as close as possible to all the points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generating synthetic data for hours studied (X) and test scores (Y)
np.random.seed(0)
X = np.random.rand(100, 1) * 10   # X values (hours studied), scaled to 0-10
Y = 3 * X + np.random.randn(100, 1) * 2   # Y values (test scores) with noise
print(X)
print(Y)
# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X, Y)

# Predict Y values using the model created above
Y_pred = model.predict(X)

# Plot the original data points and the fitted line
plt.scatter(X, Y, color='blue', label='Original data')
plt.plot(X, Y_pred, color='red', linewidth=3, label='Fitted line')

plt.title('Linear Regression Example')
plt.xlabel('Hours Studied')
plt.ylabel('Test Scores')
plt.legend()
plt.grid(True)
plt.show()
```
